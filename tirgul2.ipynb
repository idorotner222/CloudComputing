```python
# @title Python to JSON to Python

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}
print("Python Object: \n\t", x)

for key in x:
    print(key+":\t", x[key])

x_js = json.dumps(x)
print("\nCorresponding JSON Object: \n\t", x_js)

y = json.loads(x_js)
print("\nGoing back to Python: \n")
for key in y:
    print(key+":\t", y[key])
```

```python
#@title תרגיל סיכום - ניהול סטודנטים { run: "auto" }
import json
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. יצירת רשימת שמות לתיבת הבחירה מתוך הנתונים שנטענו
student_names = [f"{s['first_name']} {s['last_name']}" for s in students_data]

# 2. יצירת רכיבי הממשק (תיבות טקסט מתאימות) [cite: 436, 437, 438]
dropdown = widgets.Dropdown(options=student_names, description='בחר סטודנט:')
first_name_text = widgets.Text(description='שם פרטי:', disabled=True)
last_name_text = widgets.Text(description='שם משפחה:', disabled=True)
email_text = widgets.Text(description='מייל:', disabled=True)
show_text = widgets.Text(description='תוכנית אהובה:')
update_button = widgets.Button(description="עדכן ב-JSON", button_style='success')
output = widgets.Output() # רכיב להצגת הודעות הצלחה בצורה נקייה

# 3. פונקציה לעדכון התיבות ברגע שבוחרים סטודנט
def on_student_change(change):
    with output:
        clear_output() # מנקה הודעות קודמות בבחירת סטודנט חדש
        selected_name = change['new']
        # תיקון: השוואה נכונה בין המחרוזת לאובייקט
        student = next(s for s in students_data if f"{s['first_name']} {s['last_name']}" == selected_name)

        # עדכון אוטומטי של התיבות (Deserialization) [cite: 262]
        first_name_text.value = student['first_name']
        last_name_text.value = student['last_name']
        email_text.value = student['email']
        show_text.value = student.get('favorite_show', "")

# 4. פונקציית שמירה לקובץ (Serialization) [cite: 255, 335, 439]
def on_update_clicked(b):
    with output:
        clear_output()
        selected_name = dropdown.value
        student = next(s for s in students_data if f"{s['first_name']} {s['last_name']}" == selected_name)

        # עדכון המילון (Update) [cite: 335]
        student['favorite_show'] = show_text.value

        # שמירה פיזית לקובץ (Serialization - dump) [cite: 23, 255]
        with open(FILE_PATH, 'w', encoding='utf-8') as f:
            json.dump(students_data, f, indent=4, ensure_ascii=False)
        print(f"✅ עודכן בהצלחה עבור {selected_name}")

# חיבור האירועים
dropdown.observe(on_student_change, names='value')
update_button.on_click(on_update_clicked)

# הצגת הממשק
display(dropdown, first_name_text, last_name_text, email_text, show_text, update_button, output)

# הרצה ראשונית להצגת הנתונים של הסטודנט הראשון שמופיע
on_student_change({'new': dropdown.value})
```
