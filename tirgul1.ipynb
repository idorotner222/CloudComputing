```python
# @title שלב 4: עדכון תוכנית אהובה (Append)
# @markdown הריץ תא זה כדי להוסיף את התוכנית האהובה לסטודנט שנבחר.

# פרמטרים מהטופס
favorite_show = "האלופה" # @param {type:"string"}
student_name = "Ido Rotner" # @param {type:"string"}

# הגדרת נתיב הקובץ (זהה לשלבים הקודמים)
file_path = '/content/drive/MyDrive/אורט בראודה - הנדסת תוכנה/cloud computing/students.txt'

# שימוש במצב 'a' כדי להוסיף לסוף הקובץ
# הוספנו "\n" בתחילת המחרוזת כדי להבטיח שהעדכון יירשם בשורה חדשה
with open(file_path, 'a', encoding='utf-8') as file:
    content_to_append = f"\n{student_name}, תוכנית אהובה: {favorite_show}"
    file.write(content_to_append)

print(f"✅ הפרטים נוספו בהצלחה לסוף הקובץ!")
print(f"שורת העדכון: {content_to_append}")
```
